## Transforming data using a function or mapping

```python
data = DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami',
                           'corned beef', 'Bacon', 'pastrami', 'honey ham',
                           'nova lox'],
                  'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
  'bacon': 'pig',
  'pulled pork': 'pig',
  'pastrami': 'cow',
  'corned beef': 'cow',
  'honey ham': 'pig',
  'nova lox': 'salmon'
}
```

```python
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
data
```

```python
data['food'].map(lambda x: meat_to_animal[x.lower()])
```

## Discretization and binning

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)
cats
```

```python
cats.codes
```

```python
cats.value_counts()
```

```python
pd.cut(ages, bins, right = False)
```

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels=group_names)
```

```python
data = np.random.rand(20)
pd.cut(data, 4, precision = 2)
```

```python
pd.cut(data, 4, precision = 2).value_counts()
```

```python
data = np.random.randn(1000)
cats = pd.qcut(data, 4)
cats
```

```python
cats.value_counts()
```

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.])
```

## Detecting and filtering outliers

```python
np.random.seed(12345)
data = DataFrame(np.random.randn(1000, 4))
data.describe()
```

```python
col = data[3]
col[np.abs(col) > 3]
```

```python
data[(np.abs(data) > 3).any(1)]
```

```python
data = np.where(np.abs(data) > 3, np.sign(data) * 3, data)
data = DataFrame(data)
data.describe()
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('movies.dat', sep='::', header=None,
                        names=mnames, engine = 'python')
movies[:10]
```

```python
genre_iter = (set(x.split('|')) for x in movies.genres)
print(type(genre_iter))
genres = sorted(set.union(*genre_iter))
genres
```

```python
dummies = DataFrame(np.zeros((len(movies), len(genres))).astype(np.int32), columns = genres)
dummies
```

```python
for i, gen in enumerate(movies.genres):
    dummies.ix[i, gen.split('|')] = 1
dummies
```

```python
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.ix[0]
```
